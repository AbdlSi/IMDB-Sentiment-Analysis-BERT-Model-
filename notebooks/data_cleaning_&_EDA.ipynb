```python
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re


df = pd.read_csv("https://raw.githubusercontent.com/AbdlSi/IMDB-Sentiment-Analysis-BERT-Model-/refs/heads/main/data/raw/imdb.csv")
```

```python
df.head()
```

```python
x1 = len(df[df['sentiment'] == "positive"])
x2 = len(df[df['sentiment'] == "negative"])


categories = ["Positive","Negative"]
plt.bar(categories, height=[x1,x2])
plt.title(" ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()
```

```python
word_count = df["review"].str.split().str.len()
```

```python
plt.hist(word_count , bins =50)
plt.title(" ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()
```

```python

df["review"] = df["review"].str.replace(r'http\S+|www.\S+',"",regex = True)
df["review"] = df["review"].str.replace(r'@\S+',"",regex = True)
df["review"] = df["review"].str.replace(r'(<br\s*/?>)',"",regex = True)

```

```python
df.to_csv("/content/clean_data.csv", index=False)


files.download("/content/clean_data.csv")
```
